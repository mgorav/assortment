
## Operations Research (OR) Based Optimization for Assortment Planning

Operations research techniques can optimize the product mix to maximize profits while considering constraints such as production capacity and budget. This example demonstrates using linear programming to determine the optimal quantity of products to produce, highlighting OR's role in strategic assortment planning.


````python

from scipy.optimize import linprog

# Example of OR-based optimization for product mix
# Objective: Maximize profits based on selling prices and production costs

# Coefficients of the objective function (Profit per unit for each product)
c = [-3, -5]  # Negative values for maximization problem

# Inequality constraints (Production capacity constraints)
A = [[1, 0], [0, 2], [1, 1]]
b = [100, 100, 160]  # Maximum capacity for resources

# Solving the linear programming problem
res = linprog(c, A_ub=A, b_ub=b, bounds=(0, None))

# Displaying the optimal number of units to produce for each product
print(f'Optimal production: {res.x}')
print(f'Maximum Profit: {-res.fun}')
````


## Google OR-Tools for Retail Optimization

Google's OR-Tools provide a powerful suite of software for optimization problems, including linear programming, integer programming, and more. This section demonstrates how Google OR-Tools can be used to solve a retail optimization problem, such as determining the optimal product mix to maximize profits under specific constraints like material and labor availability.


````python

from ortools.linear_solver import pywraplp

# Create the solver using the SCIP backend
solver = pywraplp.Solver.CreateSolver('SCIP')

# Example problem: Maximize profit for product mix considering limited resources (e.g., material, labor)

# Decision variables: Number of units to produce for each product
x1 = solver.IntVar(0, solver.infinity(), 'x1')  # Product 1
x2 = solver.IntVar(0, solver.infinity(), 'x2')  # Product 2

# Objective function: Maximize total profit = 5*x1 + 10*x2
solver.Maximize(5 * x1 + 10 * x2)

# Constraints
# Material constraint: 3*x1 + 2*x2 <= 100
solver.Add(3 * x1 + 2 * x2 <= 100)

# Labor constraint: 2*x1 + 3*x2 <= 90
solver.Add(2 * x1 + 3 * x2 <= 90)

# Solve the problem
status = solver.Solve()

if status == pywraplp.Solver.OPTIMAL:
    print('Solution:')
    print('Objective value =', solver.Objective().Value())
    print('x1 =', x1.solution_value())
    print('x2 =', x2.solution_value())
else:
    print('The problem does not have an optimal solution.')
````


## API Contract: Connecting Everything End-to-End

To seamlessly integrate forecasting, optimization models, and assortment planning, a well-defined API contract is crucial. This contract specifies the interaction between various components of retail company's Assortment 2.0 initiative, ensuring flexibility and modularity in the system architecture.

### API Contract Specification

The API contract outlines methods for demand forecasting, allocation optimization, assortment definition, and ABB classification. Each method accepts specific input parameters and returns structured JSON responses, facilitating clear communication between different system components.

#### Key API Methods:

- `get_forecast`: Generates a demand forecast for a product at a specified store.
- `get_allocation`: Determines optimized allocation quantities across stores.
- `get_assortment`: Defines the optimal product assortment for a store based on customer segments.
- `get_abb`: Classifies products into Assortment Building Blocks based on attributes.

### Implementation Strategy

- **Baseline Model**: Initially, simple statistical models serve as the baseline, conforming to the API contract.
- **Validation**: Accuracy is monitored over a period, and the best-performing models are integrated into the production environment.
- **Flexibility**: The contract allows for easy swapping of models based on performance, ensuring the system remains adaptive and scalable.

This API contract plays a pivotal role in orchestrating a harmonious blend of data-driven insights and fashion expertise, revolutionizing retail planning and execution.


````python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.cluster import KMeans
import json
import warnings

# Seed for reproducibility
np.random.seed(42)

# Generate mock sales data
num_records = 50000
store_ids = ['A' + str(i) for i in range(1, 11)]  # Stores A1 to A10
product_ids = [str(i) for i in range(1, 101)]  # Products 1 to 100
dates = pd.date_range(start='2024-01-01', periods=num_records // len(product_ids) * len(store_ids), freq='D')
sales = np.random.randint(1, 100, size=num_records)

sales_data_example = pd.DataFrame({
    'store_id': np.random.choice(store_ids, num_records),
    'product_id': np.random.choice(product_ids, num_records),
    'date': np.random.choice(dates, num_records),
    'sales': sales
}).sort_values(by=['date', 'store_id', 'product_id']).reset_index(drop=True)

class AssortmentPlanningAPI:
    def __init__(self):
        self.models = {}

    def train_arima_model(self, sales_data, order=(1, 1, 1)):
        warnings.filterwarnings("ignore")
        sales_data_grouped = sales_data.groupby(['product_id', 'date']).agg({'sales': 'sum'}).reset_index()
        for product_id in sales_data_grouped['product_id'].unique():
            model_data = sales_data_grouped[sales_data_grouped['product_id'] == product_id]
            model_data = model_data.set_index('date').sales.asfreq('D').fillna(0)
            if len(model_data) < 30:
                print(f"Skipping ARIMA for {product_id}, not enough data.")
                continue
            model = ARIMA(model_data, order=order)
            self.models[product_id] = model.fit()
        warnings.resetwarnings()

    def get_forecast(self, product_id, start_date, end_date):
        if product_id not in self.models:
            return {"error": "Model not trained for product_id: " + product_id}
        forecast = self.models[product_id].get_forecast(steps=(pd.to_datetime(end_date) - pd.to_datetime(start_date)).days + 1)
        return {
            'product_id': product_id,
            'date': [pd.Timestamp(ts).strftime('%Y-%m-%d') for ts in pd.date_range(start=start_date, end=end_date)],
            'forecast_qty': forecast.predicted_mean.tolist()
        }

    def get_allocation(self, product_catalog, store_list, sales_data):
        kmeans = KMeans(n_clusters=len(store_list), random_state=42)
        features = sales_data[['product_id', 'sales']].groupby('product_id').mean().reset_index()
        kmeans.fit(features[['sales']])
        allocation = []
        for idx, cluster in enumerate(kmeans.labels_):
            allocated_products = features['product_id'][kmeans.labels_ == cluster].tolist()
            for pid in allocated_products:
                allocation.append({'product_id': pid, 'store_id': store_ids[idx % len(store_ids)], 'allocated_qty': np.random.randint(20, 50)})
        return allocation

    def get_assortment(self, store_id, customer_segments, sales_data):
        kmeans = KMeans(n_clusters=len(customer_segments), random_state=42)
        features = sales_data[['product_id', 'sales']].groupby('product_id').mean().reset_index()
        kmeans.fit(features[['sales']])
        segment_id = customer_segments.index('Adults')  # Example segment
        assortment_products = features['product_id'][kmeans.labels_ == segment_id].tolist()
        return {
            'store_id': store_id,
            'customer_segment': 'Adults',
            'product_list': assortment_products
        }

    def get_abb(self, product_id, attributes_dict):
        abb_name = attributes_dict.get('Category', 'Unknown')  # Example attribute
        return {
            'product_id': product_id,
            'abb_name': abb_name
        }

api = AssortmentPlanningAPI()
api.train_arima_model(sales_data_example)

forecast = api.get_forecast('1', '2024-01-01', '2024-01-07')
allocation = api.get_allocation(product_ids, store_ids, sales_data_example)
assortment = api.get_assortment('A1', ['Adults', 'Teens'], sales_data_example)
abb = api.get_abb('1', {'Category': 'Casual Wear'})

# Custom JSON handler to convert non-serializable objects
def custom_json_handler(obj):
    if isinstance(obj, (pd.Timestamp, np.datetime64)):
        return obj.isoformat()
    raise TypeError("Object of type '%s' is not JSON serializable" % type(obj).__name__)

# Pretty print the JSON outputs
print(json.dumps(forecast, indent=4, default=custom_json_handler))
print(json.dumps(allocation, indent=4, default=custom_json_handler))
print(json.dumps(assortment, indent=4, default=custom_json_handler))
print(json.dumps(abb, indent=4, default=custom_json_handler))
````


## Monte Carlo What-If Analysis for Assortment Planning

Monte Carlo simulation is a powerful tool for performing what-if analysis, especially in the context of retail assortment planning. By simulating thousands of scenarios using random variables, it helps in understanding the impact of various factors on sales, inventory levels, and overall profitability. This approach is invaluable for assessing the robustness of assortment plans against uncertainties in demand, supply chain disruptions, and market trends.

### Application in Assortment Planning

In assortment planning, Monte Carlo simulations can be used to:

- **Forecast Demand Variability**: Simulate different demand scenarios for new product introductions or seasonal collections.
- **Optimize Inventory Levels**: Assess the probability of stockouts or overstock conditions under varying demand and supply conditions.
- **Evaluate Promotional Strategies**: Understand the impact of different promotional activities on sales and inventory turnover.

By integrating Monte Carlo simulations, retailers can make data-informed decisions, enhancing their agility and responsiveness to market changes.


### Monte Carlo Simulations - mathematical foundation

**Monte Carlo simulations** refer to computational algorithms that rely on repeated random sampling to obtain numerical estimates of metrics. They allow modeling uncertainty and forecasting outcomes based on probability distributions of inputs.

### Retail Assortment Application 

In retail assortment planning, key applications of Monte Carlo simulation include:

- **Demand forecasting**: Simulate demand across stores by random sampling from probabilistic demand distributions parameterized based on past sales data.

- **Financial modeling**: Build models for revenue, costs, and profitability metrics under demand uncertainty to forecast financial performance.

### Mathematical Overview

The Monte Carlo approach involves:

- Define a **parametric model** based on distributions like normal, Poisson etc. For example, to model demand ($D$): 
    ```
    D ~ N(, ) 
    ```
    where  and  are estimated from data.
    
- **Randomly sample** the distribution N times to generate demand scenarios:
    ```
    Di ~ N(, ),  for i = 1 to N  
    ```

- **Evaluate model** with the random inputs over multiple simulations:

    ``` 
    Revenuei = Di * Selling Price 
    Profiti = Revenuei - Di * Unit Cost
    ```

- **Aggregate results** across simulations to analyze value distributions.

### Implementation Insights

The Python simulation code demonstrates key aspects:

- Sample demand distribution using NumPy random generation
- Parameterize inputs like price, cost for financial calculations 
- Wrap computation in loop over stores and simulation runs
- Store and analyze resulting DataFrame of simulated outputs
- Visualize demand distribution across stores 

Here is the code snippet formatted for easy copy-pasting into a Markdown file:

### Simulate Assortment Demand Function 

The `simulate_assortment_demand()` function is declared with several parameters:

- `num_stores`: Number of stores for which to simulate demand.  
- `seasons`, `holidays`, `categories`: Passed in but not used here. Could adjust demand simulation based on these factors.
- `product_price`: Price at which the product is sold. 
- `cost_price`: Cost of the product for the seller.
- `mean_demand`: The mean of the normally distributed demand.  
- `demand_std_dev`: The standard deviation of the normally distributed demand.
- `num_sims_per_store`: Number of simulations to run per store. Default is 500.

An empty DataFrame is initialized to store the results of the simulations.  

A loop runs over the number of stores. For each store:

- Simulates demand using NumPy random normal distribution with passed mean and standard deviation
- Calculates sales and profit metrics based on simulated demand
- Stores the simulated data in a DataFrame

After all stores are simulated, total number of simulations is printed.

Finally, the demand distribution per store is visualized using Seaborn boxplot and stripplot.

Monte Carlo techniques bring analytical rigor, quantification of uncertainty, and critical insights into predicting business metrics and financial performance.

````python
import seaborn as sns

def simulate_assortment_demand(num_stores, seasons, holidays, categories,
                               product_price, cost_price, mean_demand, demand_std_dev,
                               num_sims_per_store=500):

    # Initialize dataframe                  
    sim_results_df = pd.DataFrame(columns=["Store", "Demand", "Sales", "Profit"])

    # Run simulations
    for store in range(1, num_stores+1):

        store_name = f"Store {store}"

        # Simulate demand
        demand = np.random.normal(mean_demand, demand_std_dev,
                                  size=num_sims_per_store)

        # Calculate metrics
        sales = demand * product_price
        profits = sales - (demand * cost_price)

        # Store results 
        sim_data = {"Store": [store_name]*num_sims_per_store,
                    "Demand": demand,
                    "Sales": sales,
                    "Profit": profits}

        store_df = pd.DataFrame(sim_data)
        sim_results_df = pd.concat([sim_results_df, store_df])

    # Analysis
    print(f"Total Simulations: {len(sim_results_df)}")

    # Plot demand distribution  
    fig, ax = plt.subplots(figsize=(10, 8))
    sns.boxplot(data=sim_results_df, x="Store", y="Demand")
    sns.stripplot(data=sim_results_df, x="Store", y="Demand",
                  size=4, color=".3", linewidth=0)
    plt.ylabel("Simulated Demand")
    plt.title("Store-wise Demand Distribution")

    return sim_results_df

# Driver code
num_stores = 20
mean_demand = 500
demand_std_dev = 100

df = simulate_assortment_demand(num_stores, 4, ["New Year"], ["Electronics"],
                                35, 15, mean_demand, demand_std_dev)
display(df)
````

````python

import numpy as np

# Function to simulate demand for a new product launch
def simulate_product_launch(num_simulations=1000, initial_demand=100, growth_rate=0.1, volatility=0.2):
    # Simulate demand growth and variability over a period
    demand = [initial_demand]
    for _ in range(1, num_simulations):
        next_demand = demand[-1] * np.exp(growth_rate - (volatility ** 2) / 2 + volatility * np.random.normal())
        demand.append(next_demand)
    return demand

# Example simulation: New product launch over 1000 days
simulated_demands = simulate_product_launch()
average_demand = np.mean(simulated_demands)
print(f"Average simulated demand over the period: {average_demand:.2f}")

# Plotting the simulated demand scenarios
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(simulated_demands, label='Simulated Demand')
plt.title('Simulated Demand for New Product Launch')
plt.xlabel('Time (Days)')
plt.ylabel('Demand')
plt.legend()
plt.show()
````
